### model training

```python
from ultralytics import YOLO
```

```python
model = YOLO('yolov8n-cls.pt')  # load a pretrained model (recommended for training)
```

```python
results = model.train(data= r'C:\Users\US593\Work\Datasets\Water_Bottle_Classification\data', epochs=20, imgsz=64)
```

### model evaluation

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
results_path = r'C:\Users\US593\AI\04_CV\01_CV\Image Classification\Projects\Project 2\runs\classify\train\results.csv'

results = pd.read_csv(results_path)
```

```python
results.columns
```

```python
results
```

```python
plt.figure(figsize=(8,5))
plt.plot(results['                  epoch'], results['             train/loss'], label='train loss')
plt.plot(results['                  epoch'], results['               val/loss'], label='val loss', c='red')
plt.grid()
plt.title('Loss vs epochs')
plt.ylabel('loss')
plt.xlabel('epochs')
plt.legend()
plt.show()
```

```python
plt.figure()
plt.plot(results['                  epoch'], results['  metrics/accuracy_top1'] * 100)
plt.grid()
plt.title('Validation accuracy vs epochs')
plt.ylabel('accuracy (%)')
plt.xlabel('epochs')
plt.show()
```
